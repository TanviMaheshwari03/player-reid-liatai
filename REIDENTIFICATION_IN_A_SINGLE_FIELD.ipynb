```python
import numpy as np

def compute_centroid(box):
    """Calculate the (x, y) center of a bounding box."""
    x1, y1, x2, y2 = box
    return int((x1 + x2) / 2), int((y1 + y2) / 2)

def compute_iou(box1, box2):
    """Compute Intersection-over-Union between two boxes."""
    xA = max(box1[0], box2[0])
    yA = max(box1[1], box2[1])
    xB = min(box1[2], box2[2])
    yB = min(box1[3], box2[3])

    interArea = max(0, xB - xA) * max(0, yB - yA)
    boxAArea = (box1[2] - box1[0]) * (box1[3] - box1[1])
    boxBArea = (box2[2] - box2[0]) * (box2[3] - box2[1])

    if boxAArea + boxBArea - interArea == 0:
        return 0.0

    return interArea / float(boxAArea + boxBArea - interArea)
```

```python
import numpy as np
import math

class SimplePlayerTracker:
    def __init__(self, iou_threshold=0.3, max_lost=5, distance_threshold=50):
        self.next_id = 1
        self.tracks = {}  # {id: {'box': box, 'centroid': (x,y), 'lost': 0}}
        self.iou_threshold = iou_threshold
        self.max_lost = max_lost
        self.distance_threshold = distance_threshold

    def compute_centroid(self, box):
        x1, y1, x2, y2 = box
        return int((x1 + x2) / 2), int((y1 + y2) / 2)

    def compute_iou(self, box1, box2):
        xA = max(box1[0], box2[0])
        yA = max(box1[1], box2[1])
        xB = min(box1[2], box2[2])
        yB = min(box1[3], box2[3])
        interArea = max(0, xB - xA) * max(0, yB - yA)
        boxAArea = max(1, (box1[2] - box1[0]) * (box1[3] - box1[1]))
        boxBArea = max(1, (box2[2] - box2[0]) * (box2[3] - box2[1]))
        return interArea / float(boxAArea + boxBArea - interArea)

    def compute_distance(self, c1, c2):
        return math.hypot(c1[0] - c2[0], c1[1] - c2[1])

    def update(self, detections):
        new_tracks = {}
        used_detections = set()

        for track_id, track in self.tracks.items():
            best_match = None
            best_iou = 0
            best_dist = float('inf')
            track_box = track['box']
            track_centroid = track['centroid']

            for i, det in enumerate(detections):
                if i in used_detections:
                    continue

                det_centroid = self.compute_centroid(det)
                iou = self.compute_iou(track_box, det)
                dist = self.compute_distance(track_centroid, det_centroid)

                if iou >= self.iou_threshold and dist <= self.distance_threshold:
                    if iou > best_iou:
                        best_match = i
                        best_iou = iou
                        best_dist = dist

            if best_match is not None:
                matched_box = detections[best_match]
                new_tracks[track_id] = {
                    'box': matched_box,
                    'centroid': self.compute_centroid(matched_box),
                    'lost': 0
                }
                used_detections.add(best_match)
            else:
                # Increase lost count if no match
                if track['lost'] < self.max_lost:
                    new_tracks[track_id] = {
                        'box': track['box'],
                        'centroid': track['centroid'],
                        'lost': track['lost'] + 1
                    }

        # Register new players for unmatched detections
        for i, det in enumerate(detections):
            if i in used_detections:
                continue
            new_tracks[self.next_id] = {
                'box': det,
                'centroid': self.compute_centroid(det),
                'lost': 0
            }
            self.next_id += 1

        self.tracks = new_tracks
        return self.tracks
```

```python
from ultralytics import YOLO
import cv2

# Load model & tracker
model = YOLO("yolov11.pt")
tracker = SimplePlayerTracker()

cap = cv2.VideoCapture(video_path)
fps = cap.get(cv2.CAP_PROP_FPS)
width  = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
out = cv2.VideoWriter("output_reid.mp4", cv2.VideoWriter_fourcc(*'mp4v'), fps, (width, height))

frame_idx = 0
while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Run YOLO detection on the frame
    results = model.predict(source=frame, verbose=False)[0]
    detections = results.boxes.xyxy.cpu().numpy()

    # Update tracker
    tracked = tracker.update(detections)

    # Draw tracked players with ID
    for pid, info in tracked.items():
        x1, y1, x2, y2 = map(int, info['box'])
        cv2.rectangle(frame, (x1, y1), (x2, y2), (0,255,0), 2)
        cv2.putText(frame, f"ID {pid}", (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0,255,0), 2)

    out.write(frame)

    if frame_idx % 30 == 0:
        print(f"Processed frame {frame_idx}")
    frame_idx += 1

cap.release()
out.release()
print("✅ Final output saved as output_reid.mp4")
```
